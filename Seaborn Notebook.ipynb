* DISTRIBUTION PLOTS

```python
import seaborn as sns
%matplotlib inline  
#To see visualisations inside notebook
```

```python
#Predefined datasets of seaborn
tips = sns.load_dataset('tips')
tips.head()
```

```python
#DIST PLOT (distribution of 1 variable)
sns.distplot(tips['total_bill']) #Histogram + a KDE(Kernel density estimation)
#KDE is the sum of all gaussian curves created by a rugplot(explained later)
```

```python
sns.distplot(tips['total_bill'],kde = False,bins = 40)
```

```python
#JointPlot - combine distplots(2 variables)
sns.jointplot(x = 'total_bill',y = 'tip',data = tips) #Scatter plot is default
```

```python
sns.jointplot(x = 'total_bill',y = 'tip',data = tips,kind = 'hex') #Hexplot
```

```python
sns.jointplot(x = 'total_bill',y = 'tip',data = tips,kind = 'reg') #Regression plot
```

```python
sns.jointplot(x = 'total_bill',y = 'tip',data = tips,kind = 'kde') #Density plot
```

```python
#Paiplot - Pairwise relationships across an entire dataframe
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue = 'sex',palette = 'coolwarm')
```

```python
#Rugplots
sns.rugplot(tips['total_bill']) #dash mark for every plot on the distr.line
#A bell curve(Gaussian) is created for each dash in the rugplot,and the summation of all the bell curves is the KDE plot 
```

```python
sns.kdeplot(tips['total_bill'])
```

* CATEGORICAL PLOTS

```python
#BarPlot - Compare a numerical and a non-numerical(categorical) column
sns.barplot(x = 'sex',y = 'total_bill',data = tips)
```

```python
import numpy as np
sns.barplot(x = 'sex',y = 'total_bill',data = tips,estimator = np.std) #Instead of mean,we can do std or anything else too
```

```python
sns.countplot(x = 'sex',data = tips) #Counts no.of occurences in graph form. No y axis value required as its already chosen
```

```python
#BoxPlot
sns.boxplot(x = 'day',y = 'total_bill', data = tips)
```

```python
sns.boxplot(x = 'day',y = 'total_bill', data = tips,hue = 'smoker')
```

```python
#ViolinPLot
sns.violinplot(x = 'day',y = 'total_bill',data = tips)
```

```python
sns.violinplot(x = 'day',y = 'total_bill',data = tips,hue = 'sex')
```

```python
sns.violinplot(x = 'day',y = 'total_bill',data = tips,hue = 'sex',split = True)
```

```python
#StripPlot = Scatter plot where one variable is categorical
#sns.stripplot(x = 'day',y = 'total_bill',data = tips,hue = 'sex',split = True) split keyword has been redundant in Py3
sns.stripplot(x = 'day',y = 'total_bill',data = tips,hue = 'sex',dodge = True) 
#To get more clear and points more swayed, add jitter = True
```

```python
#SwarmPlot - Combining ideas of scatter, strip and violin plot
sns.swarmplot(x = 'day',y = 'total_bill',data = tips)
```

```python
sns.violinplot(x = 'day',y = 'total_bill',data = tips)
sns.swarmplot(x = 'day',y = 'total_bill',data = tips,color = 'black')
```

```python
#FactorPlot - Create any plot eith ease
sns.factorplot(x = 'day',y = 'total_bill',data = tips,kind = 'bar')
```

* MATRIX PLOTS (Heat Maps)

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr() #Converting to mar=trix form
```

```python
tc
```

```python
sns.heatmap(tc) #Heatmaps only work on matrices with relation btw the row and columns
```

```python
sns.heatmap(tc, annot = True,cmap = 'coolwarm')
```

```python
fp = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
fp
```

```python
sns.heatmap(fp,linecolor='white',linewidths=1)
```

```python
#clustermap clusters columns and rows based on similarity
sns.clustermap(fp,cmap = 'coolwarm')
```

```python
sns.clustermap(fp,cmap = 'coolwarm',standard_scale=1) #See the legend for the change
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
g = sns.PairGrid(iris) 
#Notice the capital P,G
g.map_diag(sns.distplot) #Fills the diagonal graphs
g.map_upper(plt.scatter) #Fills all graphs right of distplot(diagonal)
g.map_lower(sns.kdeplot) #Fills all graphs left of diagonal
```

```python
g = sns.pairplot(iris) #Not very customisable.So we used grids.
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
#FacetGrid
g = sns.FacetGrid(data=tips,col = 'time',row = 'smoker') #Get an empty grid to plot on
g.map(sns.distplot,'total_bill')
```

```python
#REGRESSION PLOTS
#lm plots(linear-model)
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,hue = 'sex',markers = ['o','v']) #hue separates by color
```

```python
#if you don't want male and female in the same plot
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,col = 'sex',row = 'time') 
#instead of hue = 'sex'.col separates by column.row separates by row
```
