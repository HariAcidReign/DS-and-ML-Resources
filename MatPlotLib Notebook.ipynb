* MATPLOTLIB

```python
import matplotlib.pyplot as plt
%matplotlib inline
#Allows to see plots created inside Jupyter
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
#Creating matplotlib plot: Functionl method, OO method
```

```python
#Functional Method
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
#Multiplots on the same canvas
plt.subplot(1,2,1) #no. of rows,columns,plot number
plt.plot(x,y,'r') # 'r' is for denoting red color

plt.subplot(1,2,2)
plt.plot(y,x,'b') 
```

```python
#OO (Object Oriented) Method
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #Left,bottom,width,height (graph wrt canvas is 10% from left and bottom)
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.set_title('Larger Plot')
axes2.set_title('Smaller Plot')
axes1.plot(x,y)
axes2.plot(y,x)
```

```python
#OO method of subplot creation
fig,axes = plt.subplots(nrows = 2,ncols = 2)
plt.tight_layout() #fixes overlapping issues of x and y values
```

```python
axes  #axes is an iterable array
```

```python
fig,axes = plt.subplots(nrows = 1,ncols = 2) 
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 1,ncols = 2) 
axes[0].plot(x,y)
axes[0].set_title('Title1')
axes[1].plot(y,x)
axes[1].set_title('Title2')
```

```python
#FIGURE SIZE AND DPI
fig = plt.figure(figsize = (7,2),dpi = 150) #3 and 2 are the width and ht of figure in inches. DPI is not complusory.
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize = (8,2)) 
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('PlotFig.jpeg',dpi = 200) #Saving into a file(dpi optional)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label = 'X Squared')
ax.plot(x,x**3,label = 'X Cubed')
#To distinguish them,we use a legend
ax.legend() #if overlap btw dia and legend, ax.legend(loc = 0) must be used to automatically find best position 
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color = 'green',lw = 5, alpha = 0.5,linestyle = '--',marker = 'o',markersize = 16,markerfacecolor = 'yellow',
       markeredgewidth = 6,markeredgecolor = 'black') 
#Color can be in hex codes.
#Can type lw or linewidth.alpha can change the transparency.Use ls or linestyle.
#Marker denotes the points
```

```python
len(x) #No. of marker points 
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color = 'green',lw =2,ls = '-')
#To take in only required part of the entire plot
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
